```python
# 1
num = int(input("학생 번호를 입력하세요 : "))

# (num - 1) // 10
# 1 ~ 10 => 0
leader_num = ((num - 1) // 10) * 10 + 1

print(f"학생 번호 : {num}")
print(f"팀장 번호 : {leader_num}")
```

```python
# 8
num = int(input("정수를 입력하세요."))

if num < 1:
    print("소수가 아닙니다.")
else:
    for i in range(2, num):
        if num % i == 0:
            print("소수가 아닙니다.")
            break
    else:
        print("소수입니다.")
```

```python
# 9
count = 0
for i in range(3, 101):
    for k in range(2, i):
        if i % k == 0:
            break
    else:
        count += 1
count
```

```python
# 11
out_space = 0
inner_space = 8

for i in range(1, 11):
    print(" " * out_space + "*" + " "* inner_space +"*")

    if i < 5:
        out_space += 1
        inner_space -= 2
    elif i > 5:
        out_space -= 1
        inner_space += 2

    
```
