```python
class Student:
    def __init__(self, name):
        self.name = name
    def print_name(self):
        print(self.name)
    def greet(self):
        print(f"{self.name}님 hello")
```

```python
# 상속을 하지 않는 방법
# 사용자 정의 데이터타입의 재활용
class Student2:
    def __init__(self, name):
        self.name = name
    def print_name(self):
        print(self.name)
    def study(self):
        print("공부한다.")
```

```python
s1 = Student("뷔")
s1.print_name()
```

```python
# 상속
class HighStudent(Student):
    def study(self):
        print("공부한다.")
```

```python
# 사용
s = HighStudent("홍길동")
s.print_name()
s.study()
s.greet()
```

```python
# 딕셔너리 함수

def average(student):
    return (student["kor"] + student["eng"]) / 2


student = {
    "name": "홍길동",
    "kor": 90,
    "eng": 85,
    "average": average
}

print(student["average"](student))

student2 = {
    "name": "김철수",
    "kor": 80,
    "eng": 95,
    "average": average
}

print(student2["average"](student2))
```

```python
class Student:
    def __init__(self, name, kor, eng):
        self.name = name
        self.kor = kor
        self.eng = eng
        self.math = 0 

    def average(self):
        return (self.kor + self.eng + self.math) / 3
    
s = Student("홍길동", 90, 85)
s.math = 88
s2 = Student("김철수", 50, 60)
s2.math = 70

print(s.average())
print(s2.average())
```

```python
# 학생이 여러명이면

# 코드가 계속 반복이 많고, 만약 평균 계산 방법이 바뀌면 모든 곳을 수정해야 한다.

s1 = {"name": "홍길동", "kor": 90, "eng": 85}
s2 = {"name": "이순신", "kor": 100, "eng": 95}
s3 = {"name": "김유신", "kor": 80, "eng": 90}

a1 = (s1["kor"] + s1["eng"]) / 2
a2 = (s2["kor"] + s2["eng"]) / 2
a3 = (s3["kor"] + s3["eng"]) / 2
```

```python
print(s1["name"], a1)
print(s2["name"], a2)
print(s3["name"], a3)
```

```python
# 클래스로 변경하면

class Student:

    def __init__(self, name, kor, eng):
        self.name = name
        self.kor = kor
        self.eng = eng

    def total(self):
        return self.kor + self.eng

    def average(self):
        return self.total() / 2


s1 = Student("홍길동", 90, 85)
print(s1.average())

s2 = Student("김철수", 80, 88)
print(s2.average())
```
