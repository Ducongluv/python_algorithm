```python
queue = ["A", "B", "C"]

pop_index = 0

while pop_index < len(queue):
    print(queue[pop_index])
    print(queue[pop_index + 1:])
    pop_index += 1

print(queue)
```

```python
# 음식 주문 예제

queue = []

queue.append("햄버거")
queue.append("피자")
queue.append("라면")

print("주문 목록 :", queue)

print("조리 시작 :", queue.pop(0))
print("조리 시작 :", queue.pop(0))

print("남은 주문 :", queue)
```

```python
# 20, 30, 40, 50이 모두 앞으로 이동하므로 데이터가 많을수록 비효율적이다.

# 그래서 실제 알고리즘에서는 큐를 구현할 때 collections모듈의 deque를 사용한다.

# deque의 popleft() 메서드를 사용하면 맨 앞의 데이터를 이동 없이 빠르게 꺼낼 수 있어 알고리즘 문제에서는 deque를 사용하는 것이 표준적인 방법
from collections import deque

queue = deque()

queue.append("A")
queue.append("B")
queue.append("C")

print(queue.popleft())
print(queue.popleft())
```
