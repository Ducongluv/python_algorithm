```python
import random
import time

# 0 ~ 9999 사이의 숫자 1000개를 무작위로 생성
numbers = random.sample(range(10000), 1000)
len(numbers)
```

```python
start = time.time()
# start 1970년 1월 1일 0시 0.0000 카운팅

# 마지막 원소를 찾도록 설정(최악의 경우 : 최대실행횟수 확인)
target = numbers[-1]
count = 0

# 처음부터 끝까지 하나씩 비교
for n in numbers:
    count += 1
    if n == target:
        break
    
# 실행 시간 측정 종료
end = time.time()

print("선형 탐색 : ", end - start, "count : ", count)
```

```python
import random
import time

# 0~9999 사이의 숫자 1000개를 무작위로 생성
numbers = random.sample(range(10000), 1000)

# ===========================
# 선형 탐색 (O(n))
# ===========================

# 실행 시간 측정 시작
start = time.time()

# 마지막 원소를 찾도록 설정(최악의 경우)
target = numbers[-1]

# 처음부터 끝까지 하나씩 비교
for n in numbers:
    if n == target:
        break

# 실행 시간 측정 종료
end = time.time()

print("선형 탐색 :", end - start)

# ===========================
# 버블 정렬 (O(n²))
# ===========================

# 새로운 무작위 데이터 생성
numbers = random.sample(range(10000), 1000)

# 실행 시간 측정 시작
start = time.time()

# 버블 정렬 수행
for j in range(len(numbers) - 1):
    for i in range(len(numbers) - 1 - j):
        if numbers[i] > numbers[i + 1]:
            numbers[i], numbers[i + 1] = numbers[i + 1], numbers[i]

# 실행 시간 측정 종료
end = time.time()

print("버블 정렬 :", end - start)
```
