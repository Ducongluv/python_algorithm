```python
# re.findall() 모든 검색 결과를 리스트로 반환한다.

import re

text = "apple banana orange"
result = re.findall(r"a", text)
print(result) # ['a', 'a', 'a', 'a', 'a']
import re

text = "A10 B20 C30"
numbers = re.findall(r"\d+", text) # \d 숫자한문자, \d+ 연속된 숫자 한문자이상
print(numbers)
```

```python
# re.sub(패턴, 변경문자, 문자열)

# 숫자 문자 제거
import re

text = "A10 B20 C30"
result = re.sub(r"\d", "", text)
result2 = re.sub(r"\D", "0", text)
print(result)
print(result2)
```

```python
# 상품명에서 가격만 출력

import re

products = [
    "사과 3500원",
    "배 4200원",
    "수박 18000원"
]

price_list = []

for product in products:
    name = re.search(r"[가-힣] +", product)
    price = re.search(r"\d+", product)
    price_list.append({name.group():int(price.group())})

print(price_list)
```

```python
# 여러 공백을 하나의 공백으로 치환

import re

text = "Python   Data   Analysis"
result = re.sub(r"\s+", " ", text) # \s+ 1개이상의 공백 문자
print(result)
```

```python
import re

emails = [
    "python@test.com",
    "kim123@gmail.com",
    "abc@company.co.kr",
    "hello",
    "@gmail.com"
]

pattern = r"^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$"

for email in emails:
    if re.fullmatch(pattern, email):
        print(email, "정상")
    else:
        print(email, "오류")
```

```python
# URL 검증

# ^https?://([\w-]+\.)+[\w-]+(/[^\s]*)?$
import re

urls = [
    "https://www.google.com",
    "http://naver.com",
    "https://python.org/docs",
    "google.com"
]

pattern = r"^https?://([\w-]+\.)+[\w-]+(/[^\s]*)?$"

for url in urls:
    if re.fullmatch(pattern, url):
        print(url, "정상")
    else:
        print(url, "오류")
```

```python
# 주민등록변호 형식 검사(외국사람 제외)

# ^\d{6}-[1-4]\d{6}$
# 숫자6시작 

# -문자 

# 1~4문자 하나 

# 숫자문자 6 끝

import re

nums = [
    "901010-1234567",
    "000101-4234567",
    "9010101234567",
    "90101-1234567"
]

pattern = r"^\d{6}-[1-4]\d{6}$"

for n in nums:
    if re.fullmatch(pattern, n):
        print(n, "형식 정상")
    else:
        print(n, "오류")
```
