## 객체의 상태 변경

```python
class Product:
    def __init__(self, name, price, stock):
        self.name = name
        self.price = price
        self.stock = stock

    def total_price(self, count):
        return self.price * count

    def print_info(self):
        print(f"상품명 : {self.name}")
        print(f"가격 : {self.price}")
        print(f"재고 : {self.stock}")

    def stock_in(self, count):
        if count > 0:
            self.stock += count
            print(f"{self.name} 입고 {self.stock}개")
        else:
            print("입고 수량은 0보다 커야 합니다.")
    def stock_out(self, count):
        if count <= 0:
            print("출고 수량은 0보다 커야 합니다.")
        elif self.stock >= count:
            self.stock -= count
            print(f"{self.name} 입고 {self.stock}개")
        else:
            print("재고가 부족합니다.")

    def change_price(self, price):
        if price >= 0:
            self.price = price
            print(f"{self.name} 가격 {self.price}원으로 변경")
        else:
            print("가격은 0보다 작을 수 없습니다.")
```

```python
# 객체 생성
keyboard = Product("키보드", 50000, 10)
mouse = Product("마우스", 30000, 20)
monitor = Product("모니터", 250000, 5)
```

```python
# 객체 변경
keyboard.stock_in(5)
mouse.stock_out(7)
monitor.change_price(370000)
```

```python
class Product:
    products = []   # 전체 상품 저장 리스트

    def __init__(self, name, price, stock):
        self.name = name
        self.price = price
        self.stock = stock

        Product.products.append(self)

    # 상품 정보 출력
    def print_info(self):
        print(f"상품명 : {self.name}")
        print(f"가격 : {self.price}원")
        print(f"재고 : {self.stock}개")
        print("----------------")

    # 주문
    def order(self, count):
        if self.stock < count:
            print("재고 부족")
        else:
            self.stock -= count
            print(f"{self.name} {count}개 주문")
            print(f"결제금액 : {self.total_price(count)}원")
            print(f"남은 재고 : {self.stock}개")

    # 주문 금액 계산
    def total_price(self, count):
        return self.price * count

    # 입고
    def stock_in(self, count):
        self.stock += count
        print(f"{self.name} {count}개 입고")

    # 출고
    def stock_out(self, count):
        if self.stock < count:
            print("재고 부족")
        else:
            self.stock -= count
            print(f"{self.name} {count}개 출고")

    # 가격 변경
    def change_price(self, price):
        if price >= 0:
            self.price = price
            print(f"{self.name} 가격 {price}원 변경")
        else:
            print("가격은 0원 이상이어야 합니다.")

    # 할인 적용
    def discount(self, percent):
        self.price = int(self.price * (100 - percent) / 100)
        print(f"{self.name} {percent}% 할인 적용 : {self.price}원")

    # 품절 확인
    def is_sold_out(self):
        return self.stock == 0


    # -------------------------
    # 전체 상품 관리 메서드
    # -------------------------

    # 전체 상품 조회
    @classmethod
    def show_products(cls):
        for p in cls.products:
            p.print_info()

    # 상품 검색
    @classmethod
    def search_product(cls, name):
        for p in cls.products:
            if p.name == name:
                return p

        print("상품 없음")
        return None

    # 품절 상품 조회
    @classmethod
    def sold_out_products(cls):
        for p in cls.products:
            if p.is_sold_out():
                p.print_info()

    # 가장 비싼 상품
    @classmethod
    def expensive_product(cls):
        result = cls.products[0]

        for p in cls.products:
            if p.price > result.price:
                result = p

        return result

    # 재고가 가장 많은 상품
    @classmethod
    def most_stock_product(cls):
        result = cls.products[0]

        for p in cls.products:
            if p.stock > result.stock:
                result = p

        return result

    # 전체 재고 수량
    @classmethod
    def total_stock(cls):
        total = 0

        for p in cls.products:
            total += p.stock

        return total

    # 전체 재고 금액
    @classmethod
    def total_stock_price(cls):
        total = 0

        for p in cls.products:
            total += p.price * p.stock

        return total
```

```python
Product("키보드", 50000, 10)
Product("마우스", 30000, 20)
Product("모니터", 250000, 5)


# 전체 조회
Product.show_products()


# 검색 후 주문
keyboard = Product.search_product("키보드")
keyboard.order(2)


# 할인
monitor = Product.search_product("모니터")
monitor.discount(10)


# 품절 조회
Product.sold_out_products()


# 가장 비싼 상품
print("가장 비싼 상품")
Product.expensive_product().print_info()


# 전체 재고
print("전체 재고 :", Product.total_stock())


# 전체 재고 금액
print("전체 재고 금액 :", Product.total_stock_price())
```
