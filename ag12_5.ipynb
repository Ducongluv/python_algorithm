```python
#  덱 생성

from collections import deque
dq = deque()
print(dq)
```

```python
# 오른쪽 데이터 추가 append
from collections import deque

dq = deque()
dq.append(10)
dq.append(20)
dq.append(30)

print(dq)
```

```python
# 왼쪽 데이터 추가 appendleft
from collections import deque

dq = deque()
dq.append(20)
dq.append(30)
dq.appendleft(10)

print(dq)
```

```python
# 오른쪽 삭제 pop

from collections import deque

dq = deque([10, 20, 30])
x = dq.pop()

print(x)
print(dq)
```

```python
# 왼쪽 삭제 popleft

from collections import deque

dq = deque([10, 20, 30])
x = dq.popleft()

print(x)
print(dq)
```

```python
# 스택이 필요 -> 효율성 deque x -> list 사용
# 큐 필요 -> 효율성 o -> deque 사용
# 덱 필요 -> deque 사용
#  자료구조의 회전 -> 양방향(양쪽 끝) 삽입, 삭제 like 시계 방향, 반시계 방향

dq = deque([1, 2, 3, 4, 5])
dq.rotate(1) # 우측으로 한칸 이동(회전)

print(dq) 

dq1 = deque([1, 2, 3, 4, 5])
x = dq1.pop()
dq1.appendleft(x)
print(dq1)
```

```python
# 최근 방분 기록 n개 저장

from collections import deque

history = deque(maxlen=5) # maxlen=n n개만 저장

history.append("google")
history.append("naver")
history.append("youtube")
history.append("github")
history.append("chatgpt")

print(history)

history.append("python")

print(history)
```

```python
# 카드 게임(원카드, 우노, ..) 시뮬레이션

from collections import deque

cards = deque([1, 2, 3, 4, 5])

while len(cards) > 1:
    print("버림 :", cards.popleft())
    cards.append(cards.popleft())
    print("남은 카드 :", cards)

print("마지막 카드 :", cards[0])
```
