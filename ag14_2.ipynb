```python
# "\\d" -> r"\d"
print("Hello\nworld\\")
# 출력
# Hello
# world\

print(r"Hello\nworld\\")
# 출력 
# Hello\nworld\\
```

```python
import re

text = "apple banana cat"
result = re.findall(r"[abc]", text) # "apple banana cat"를 0부터 끝까지 "a", "b", "c" 중 일치하면 반환되는 리스트에 추가
print(result)
# ['a', 'b', 'a', 'a', 'a', 'c', 'a']
```

```python
import re

text = "A12B34"
print(re.findall(r"[0-9]", text))
# ['1', '2', '3', '4']
# 문자0부터 문자9까지 하나라도 일치하면 반환되는 리스트에 추가 
# 정규표현식 [0-9]는 [0123456789] 동일한 패턴
```

```python
import re

text = "Python JAVA AI"
print(re.findall(r"[A-Z]", text))
```

```python
import re

text = "a7bc123"
result = re.findall(r"[^0-9]", text) # 0~9문자가 아니면 반환되는 리스트에 추가
print(result)
```

```python
import re

password = input("비밀번호 : ")

result = []

if re.search(r"[가-힣]", password):
    result.append("한글을 사용할 수 없습니다")

if not re.search(r"[a-z]", password):
    result.append("소문자가 없습니다.")

if not re.search(r"[A-Z]", password):
    result.append("대문자가 없습니다.")

if not re.search(r"[0-9]", password):
    result.append("숫자가 없습니다.")

if not re.search(r"[^a-zA-Z0-9]", password):
    result.append("특수문자가 없습니다.")

if len(result) == 0:
    result.append("사용 가능한 비밀번호입니다.")

for r in result:
    print(r)
```
