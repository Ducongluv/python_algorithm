```python
# 1. 누적합
steps = [5231, 6120, 4890, 7005, 6500, 7200, 6800]
```

```python
total = 0
for i in range(len(steps)):
    total = total + steps[i]
    print(f"{i + 1}일 누적 : {total}")
```

```python
# 2. 누적합(Prefix Sum) - 1번 응용
steps = [
    {"2026-01-01": 5231},
    {"2026-01-02": 6210},
    {"2026-01-03": 4890},
    {"2026-01-04": 7005},
    {"2026-01-05": 6500},
    {"2026-01-06": 7200},
    {"2026-01-07": 6800}
]
```

```python
total2 = 0

for day in steps:
    for date, count in day.items():
        total2 = total2 + count
        print(f"{date} 누적 : {total2}")
```

```python
# 3. 슬라이딩 윈도우
sales = [12, 15, 18, 20, 17, 25, 22, 19]
```

```python
k = 3  # 연속된 3시간

# 처음 3시간의 합을 직접 계산
window = 0
for i in range(k):
    window += sales[i]

max_sum = window

# 첫 번째 구간 출력
print(f"{sales[0]} + {sales[1]} + {sales[2]} = {window}")

# 슬라이딩 윈도우
for i in range(k, len(sales)):
    window = window - sales[i - k] + sales[i]

    print(f"{sales[i-2]} + {sales[i-1]} + {sales[i]} = {window}")

    if window > max_sum:
        max_sum = window

print()
print(f"최대 매출 합 : {max_sum}")
```

```python
# 4. 슬라이딩 윈도우 - 3번 응용
sales = [
    {"9시": 12},
    {"10시": 15},
    {"11시": 18},
    {"12시": 20},
    {"13시": 17},
    {"14시": 25},
    {"15시": 22},
    {"16시": 19},
    {"17시": 24},
    {"18시": 21},
    {"19시": 18},
    {"20시": 16},
    {"21시": 14},
    {"22시": 11}
]
```

```python
k = 3

# 처음 3시간의 합 계산
window = 0
for i in range(k):
    window += list(sales[i].values())[0]

max_sum = window
max_start = 0

# 첫 번째 구간 출력
start_time = list(sales[0].keys())[0]
end_time = list(sales[k - 1].keys())[0]
print(f"{start_time} ~ {end_time} : {window}")

# 슬라이딩 윈도우
for i in range(k, len(sales)):
    prev = list(sales[i - k].values())[0]
    curr = list(sales[i].values())[0]

    window = window - prev + curr

    start = i - k + 1
    start_time = list(sales[start].keys())[0]
    end_time = list(sales[start + k - 1].keys())[0]

    print(f"{start_time} ~ {end_time} : {window}")

    if window > max_sum:
        max_sum = window
        max_start = start

# 최대 매출 시간대 출력
start_time = list(sales[max_start].keys())[0]
end_time = list(sales[max_start + k - 1].keys())[0]

print()
print("최대 매출 합 :", max_sum)
print("시간대 :", start_time, "~", end_time)
```
