```python
scores = [ 10, 20, 30, 40, 50, 60]

k = 3 # window size

# window 시작 포인트(인덱스 0)의 합을 초기값
window = sum(scores[0:k])

# 초기값에서 1칸으로 이동 -> k 까지
window = sum(scores[0:k])

print(window)

for i in range(k, len(scores)):
    window = window - scores[i - k] + scores[i]
    print(window)
```

```python
scores = [10, 20, 3, 40, 5, 60, 8, 55, 44]

k = 3

window = sum(scores[0:k])

totals = [window]
avgs = [window/k]
max_val = {"구간" : f"0 ~ {k - 1}", "합" : window}
min_val = {"구간" : f"0 ~ {k - 1}", "합" : window}

print(window) # 60

for i in range(k, len(scores)):
    window = window + scores[i] - scores[i - k]
    totals.append(window)
    avgs.append(window/k)

    section = f"{i - (k - 1)} ~ {i}"

    if window > max_val["합"]:
        max_val = {"구간" : section, "합" : window}

    if window < min_val["합"]:
        min_val = {"구간" : section, "합" : window}


print(totals)
print(avgs)
print(max_val)
print(min_val)
```
