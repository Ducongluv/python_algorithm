## 빈도수 계산

```python
# 상품별 판매 건수

count = {} 

for r in sales:
    key_name = r["상품"]
    if key_name not in count:
        count[key_name] = 0
    else:
        count[key_name] += 1

print(count)

for k, v in count.items():
    print(f"{k}(는)은 {v}건 입니다")
```

## 그룹별 집계

```python
# 지역별 총 판매액

local_sum = {} 

for r in sales:
    price = r["판매액"]
    # price "","결측"이면 
    if not isinstance(price, int): # if pass else ... -> if continue ...
        continue

    key_name = r["지역"]

    if key_name not in local_sum:
        local_sum[key_name] = 0
    local_sum[key_name] += price

local_sum
```

## 이동평균 

```python
# 슬라이딩 윈도우에 사용될 리스트
prices = []

for r in sales:
    if isinstance(r["판매액"], int):
        prices.append(r["판매액"])
    

window = 3

for i in range(len(prices) - (window - 1)):
    total = sum(prices[i : i + window])
    avg = round(total / window, 3)
    print(f"{i + 1} ~ {i + 3} 평균 판매액 : {avg}원")
```

## 누적평균

```python
prices = []
for r in sales:
    if isinstance(r["판매액"], int):
        prices.append(r["판매액"])

prefix = [0]

for p in prices:
    prefix.append(prefix[-1] + p)

avgs = []
for i in range(1, len(prefix)):
    avgs.append(prefix[i] / i)

avgs
```

## 결측치 처리

```python
avg = sum(prices) / len(prices)

# 중앙값 정렬

prices.sort()

if len(prices) % 2 == 1: # 홀수 이면
    median = prices[len(prices) // 2]
else: # 짝수 이면
    median = (prices[len(prices) // 2] + prices[len(prices) // 2-1]) / 2

print(median)

for r in sales:
    if not isinstance(r["판매액"], int):
        r["판매액"] = median

sales
```

## 범주별 통계

```python
result = {}

for r in sales:
    p_name = r["상품"]
    price = r["판매액"]
    total_cnt = r["수량"]
    if r["상품"] not in result:
        result[p_name] = {"합계" : 0, "건수" : 0, "최대" : 0, "최소" : 0, "총수량" : 0}
    
    info = result[p_name]
    info["합계"] += price
    info["건수"] += 1
    info["총수량"] += total_cnt

    # 최대/최소
    if info["최대"] < price:
        info["최대"] = price
    
    if info["최소"] > price:
        info["최소"] = price
result
```
